```python
motcles_fulldata = pd.pivot_table(fulldata, index="Type_media", values="domaine", columns="type mot-clés", aggfunc="count")
print(motcles_fulldata)
```

Ça, c'est intéressant!
On apprend que: 
- la portion média hors-québec ne compte pas pour beaucoup. 
- Pour les mots-clés locaux, ce sont les médias locaux qui sont les plus fréquents
- Pour les mots-clés mixtes, c'est plutôt égale, comme anticipé
- Pour les mots-clés nationaux, ce sont les termes requête qui prime mais rien de surprenant là encore.
- Il y a aussi plus de résultats pour les mots-clés local que national ou mixte, mais c'est normal car plus de mots-clés à la base. 

```python
ville_type_fulldata = pd.pivot_table(fulldata, values="domaine", aggfunc="count", index=["Villes", "type mot-clés"], columns="Type_media")
ville_type_fulldata.to_csv("villesparmotscles.csv", na_rep="0.0")
```
