# Coefficient de Gini

```python
## Pivot table qui comprend le nombre de médias le type_media par type de mot-clés par ville ##
type_media_by_ville_by_type_motsclés2 = pd.pivot_table(fulldata, values="domaine", aggfunc="count", columns=["Villes", "TypeMotClés"], index="TypeMedia", fill_value=0.0)
type_media_by_ville_by_type_motsclés2
```

```python
def gini(x):
    total = 0
    for i, xi in enumerate(x[:-1], 1):
        total += np.sum(np.abs(xi - x[i:]))
    return total / (len(x)**2 * np.mean(x))
```
