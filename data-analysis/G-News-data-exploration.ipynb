# Création de CSV et Tableau croisée dynamique pour explorer davantage

```python
## Tableau Villes par type_mots-clés 
villes_by_type_motsclés = pd.pivot_table(fulldata, index="Villes", values="domaine", columns="type mot-clés", aggfunc="count")
villes_by_type_motsclés
```

```python
## Tableau Villes par mots-clés
villes_by_motsclés = pd.pivot_table(fulldata, index="Villes", values="domaine", columns="Mots-clés", aggfunc="count")
villes_by_motsclés.to_csv("villes_by_motsclés.csv", na_rep="0.0")
```

```python
## Détail du Type_media selon le type mot-clés
motcles_by_media_type = pd.pivot_table(fulldata, index="Type_media", values="domaine", columns="type mot-clés", aggfunc="count")
motcles_by_media_type
```

Ça, c'est intéressant!
On apprend que: 
- la portion média hors-québec ne compte pas pour beaucoup. 
- Pour les mots-clés locaux, mixtes et nationaux, ce sont les médias locaux qui sont les plus fréquents.
    - Cela pourrait s'expliquer par le fai que Radio-Canada compte divers chapeau et est le média le plus populaire. 
- Il y a aussi plus de résultats pour les mots-clés local que national ou mixte, mais c'est normal car plus de mots-clés à la base. 

```python
## CSV qui comprend le nombre de médias le type_media par type de mot-clés par ville ##
type_media_by_ville_by_type_motsclés = pd.pivot_table(fulldata, values="domaine", aggfunc="count", index=["Villes", "type mot-clés"], columns="Type_media")
type_media_by_ville_by_type_motsclés.to_csv("villesparmotscles2.csv", na_rep="0.0")
```

Document généré, dispo dans le dossier data-analysis. C'est plutôt intéressant.
